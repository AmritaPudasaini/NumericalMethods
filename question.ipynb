## For f(x)=7e^0.5x at x=2 with varying step size, h=0.3, 0.15, 0.10, 0.01, 0.001

```python
import numpy as np

def f(x):
    return 7 * np.exp(0.5 * x)

x = 2

h_values = [0.3, 0.15, 0.10, 0.01, 0.001]

def exact_derivative(x):
    return 7 * 0.5 * np.exp(0.5 * x)

exact = exact_derivative(x)

print(f"{'h':>8} {'Approx Derivative':>20} {'Error':>15}")
for h in h_values:
    approx = (f(x + h) - f(x)) / h
    error = abs(approx - exact)
    print(f"{h:8.5f} {approx:20.10f} {error:15.10f}")
```

## Write a Python program to display the following result.

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(0,10,50)
y1=2*np.sin(x)
y2=np.cos(x)
fig, (ax1, ax2) = plt.subplots(2,1,figsize=(8,6))

ax1.plot (x,y1, 'o-', color='black')
ax1.set_ylabel("2sin(x)")
ax1.grid(True)

ax2.plot (x,y2, '^-', color='black')
ax2.set_ylabel("cos(x)")
ax2.set_xlabel("x")
ax2.grid(True)

plt.show()
```

## Find the roots of y=2x^2 + 5x - 25

```python
import math

a = 2
b = 5
c = -25

D = b**2 - 4*a*c

if D >= 0:
    root1 = (-b + math.sqrt(D)) / (2*a)
    root2 = (-b - math.sqrt(D)) / (2*a)
    print("Root 1:", root1)
    print("Root 2:", root2)
else:
    print("No real roots")
```

## Find equation of line passing through A(1,2), B(5,9).

```python
x1, y1 = 1, 2
x2, y2 = 5, 9

m = (y2 - y1) / (x2 - x1)

c = y1 - m * x1

print(f"Equation of line: y = {m}x + {c}")
```

## Solve equations
### 2x-3y=7
### 3x-6y=10

```python
import numpy as np

# Coefficient matrix A and constant vector B
A = np.array([[2, -3],
              [3, -6]])
B = np.array([7, 10])

# Check if the system has a unique solution
det_A = np.linalg.det(A)
if det_A != 0:
    solution = np.linalg.solve(A, B)
    x, y = solution
    print(f"Unique solution: x = {x}, y = {y}")
else:
    rank_A = np.linalg.matrix_rank(A)
    augmented = np.column_stack((A, B))
    rank_augmented = np.linalg.matrix_rank(augmented)

    if rank_A == rank_augmented:
        print("The system has infinitely many solutions.")
    else:
        print("The system has no solution.")
```
